```python
import pandas as pd
import json
import numpy as np
import os
from pathlib import Path
from collections import defaultdict
```

```python
dss_dir = Path('/home/data')
```

```python
ds_dic = defaultdict(list)
for ds_dir in os.listdir(dss_dir):
    json_path = dss_dir / ds_dir / 'info.json'
    with open(json_path, 'r') as f:
        data = json.load(f)
    for key in ['name', 'task_type', 'train_size', 'n_num_features', 'n_bin_features']:
        if key in data:
            ds_dic[key].append(data[key])
        else:
            ds_dic[key].append(None)
    ds_dic['dir'].append(ds_dir)
    
```

```python
df = pd.DataFrame.from_dict(ds_dic)
```

# These are valid datasets for our task
